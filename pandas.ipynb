```python
import pandas as pd
srs=pd.Series([1,2,3,12.2,23,32.3],index=['India','china','USA','Russia','Brazil','Sri Lanka'])
srs.name="Growth rate"
srs.index.name="Country"
srs.value="rate"
print(srs)

```

```python
#Series in python

srs1=pd.Series([1,2,3,12.2,23,32.3],index=['India','china','USA','Russia','Brazil','Sri Lanka'])
srs2=pd.Series([11,32.3,33,122,233,323.3],index=['africa','Sri Lanka','brazil','USA','bangladesh','england'])

srs=srs2/srs1

srs.name="Res"
srs.index.name="Country"
print(srs)

#is null

print(pd.isnull(srs))

#not null

print(pd.notnull(srs))
```

```python
#Indexes that satisfied sub condition
print(srs[srs==1])
#Indexes that satisfied sub condition
print(srs[srs!=1])
```

```python
#Indexing methodes in pandas

srs1=pd.Series([1,2,3,4,5],index=['A','B','C','D','E'])
srs1.name='Growth'
srs1.index.name='names'
srs2=srs1.reindex(['A','B1','C1','D','E1','F1'])
srs3=srs1.reindex(['A1','B','C1','D1','E1','F1'],fill_value=0)
print(srs1)
print(srs2.ffill())
print(srs3)
```

```python
#arrays in DataFrames
arr2=np.arange(16).reshape(4,4)
df=pd.DataFrame(arr2,index=['r1','r2','r3','r4'],columns=['c1','c2','c3','c4'])
print(df)
```
